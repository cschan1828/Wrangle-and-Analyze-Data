### Define:
**The name column has many entries (about 109) which do not look like names**

### Code:
    This function resposable extract the names for spilt text inside df column and matched usining re function 
    we will be pass the dataset and test the text column and matched using re tool
    finally this function return extracted names from text coumn

```python
import re
df_arch_clean['txt_splt'] = df_arch_clean['text'].str.split()
all = []

def check_relly_name(df):  

    if 'named' in df['text'] and re.match(r'[A-Z].*', df['txt_splt'][(df['txt_splt'].index('named') + 1)]): 
            all.append(df['txt_splt'][(df['txt_splt'].index('named') + 1)])  
    elif df['text'].startswith('Here is a ') and re.match(r'[A-Z].*', df['txt_splt'][3]):
            all.append(df['txt_splt'][3].strip('.').strip(','))         
    elif df['text'].startswith('Here we have ') and re.match(r'[A-Z].*', df['txt_splt'][3]):
            all.append(df['txt_splt'][3].strip('.').strip(',')) 
    elif df['text'].startswith('This is ') and re.match(r'[A-Z].*', df['txt_splt'][2]):
            all.append(df['txt_splt'][2].strip('.').strip(',')) 
    elif df['text'].startswith('Say hello to ') and re.match(r'[A-Z].*', df['txt_splt'][3]):
            all.append(df['txt_splt'][3].strip('.').strip(',')) 
    elif df['text'].startswith('Meet ') and re.match(r'[A-Z].*', df['txt_splt'][1]):
            all.append(df['txt_splt'][1].strip('.').strip(',')) 
    else:
        all.append('Not_Provide') 
        
df_arch_clean.apply(check_relly_name, axis=1)

df_arch_clean['Extracted_name'] = all
```

# Analysis

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('twitter_archive_master.csv')
```

## The relationship between Retweets and Favorites

```python
df['favorite_count'].plot(y='retweet_count')
plt.title('Favorites and Retweets ')
plt.xlabel('Favorites')
plt.ylabel('Retweets');
```

## The relationship between Retweets and Favorites with Rating over the time

```python
df.plot(x='rating_numerator', y='retweet_count', kind='scatter')
plt.title('Retweets and Favorites Relationship with Rating ')
plt.ylabel('Retweets')
plt.xlabel('Rating');
```

## How the Rating increase over the time

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.set_index('timestamp', inplace=True)
```

```python
df.plot(y ='rating_numerator', ylim=[0,14], style = '.', alpha = 0.4)
plt.title('Rating with Time')
plt.xlabel('Date')
plt.ylabel('Rating');
```

```python
df[['favorite_count', 'rating_numerator', 'retweet_count']].corr(method='pearson')
```

### Dog Stages Stats

```python
df.plot(column='rating_numerator', by='dog_type', kind ='bar');
plt.title('Dog type count')
plt.xlabel('Dogs Type')
plt.xlabel('Rating');
```

```python
df.groupby('dog_type')['rating_numerator'].describe()
```
